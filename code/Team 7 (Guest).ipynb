```python
# Association Rule
nAirbnb = dfStr.values.tolist()
```

```python
te = TransactionEncoder()
te_ary = te.fit(nAirbnb).transform(nAirbnb)
te_ary.astype("int")
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

frequent_itemsets
```

```python
#데이터 정제-Null샘플이 있는지 확인
print(data.isnull().sum())
```

```python
#Null 값을 가진 샘플 제거하고 제거수 확인
data.dropna(axis=0, inplace=True)
print('전체 샘플수 :', (len(data)))
```

```python
#예측변수 행렬
X = data[['Number of reviews', 'Reviews per month', 
          'Review Scores Accuracy', 'Review Scores Cleanliness', 
          'Review Scores Checkin', 'Review Scores Communication', 
          'Review Scores Location', 'Review Scores Value']]
```

```python
#독립변수를 나타내는 벡터
y = data[['Review Scores Rating']] #Review Scores Rating
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
#주성분 분석을 수행
model = PCA() #주성분 분석 계급의 인스턴스 생성
results = model.fit(X) #변수의 수를 줄일 수 있는지 알기 위해 예측변수에 PCA를 적용
Z = results.transform(X) #결과를 배열로 변환해 데이터를 새로 생성하는데 사용할 수 있게 한다
plt.plot(results.explained_variance_) #변수들의 변량을 설명하는 플롯(산비탈그림)
plt.show() #플롯을 나타냄
```

```python
pd.DataFrame(results.components_, columns=list(
    ['Number of reviews', 'Reviews per month', 
    'Review Scores Accuracy', 'Review Scores Cleanliness', 
    'Review Scores Checkin', 'Review Scores Communication', 
    'Review Scores Location', 'Review Scores Value']))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
#주성분 분석 이전의 Review Scores Rating 점수 예측
gnb = GaussianNB() #Use Gaussian Distribution naive_bayes for Estimation
fit = gnb.fit(X,y)
pred = fit.predict(X) #Predict unseen data
print(confusion_matrix(pred,y))#혼돈행렬분석
print(confusion_matrix(pred,y).trace())
```

```python
#주성분 수를 늘림으로써 Review Scores Rating 예측
predicted_correct = []#배열은 올바르게 관측된 관측값으로 채워짐
for i in range(1,7): #Loop for the first seven main components detected
    model = PCA(n_components = i)#1개 성분(첫 이터레이션)에서 7개 성분(7번째 이터레이션)을 가지고 PCA 모델을 인스턴스화한다
    results = model.fit(X)#PCA모델을 x변수(특성)에 적합하게 한다
    Z = results.transform(X) #Z는 행렬 형태의 결과(실제로는 배열이 들어있는 배열)
    fit = gnb.fit(Z,y)#추정을 위해 가우스분포 나이브베이즈분류기 사용
    pred = fit.predict(Z)#적합화된 모델을 사용한 실제 예측 자체
    predicted_correct.append(confusion_matrix(pred,y).trace())#각 이터레이션 끝에서 올바르게 분류된 관측값의 개수를 더한다
    print(predicted_correct)#이 배열을 프린트함으로써 각 이터레이션 후에 올바르게 분류된 관측이 추가된 새로운 카운트가 어떻게 되는지 볼 수 있음
    
plt.plot(predicted_correct) #보기 쉽게 배열을 그림으로
plt.show()#그림표시
```
